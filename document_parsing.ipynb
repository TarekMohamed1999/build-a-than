```python
import zipfile
import xml.etree.ElementTree as ET
```

```python
doc = zipfile.ZipFile('FGAN-O-013-R1.docx').read('word/document.xml')
root = ET.fromstring(doc)
```

```python
ns = {'w': 'http://schemas.openxmlformats.org/wordprocessingml/2006/main'}
body = root.find('w:body', ns)  # find the XML "body" tag
p_sections = body.findall('w:p', ns) 
```

```python
from docx import Document
```

```python
document = Document("FGAN-O-013-R1.docx")
```

```python
document.paragraphs[80].text
```

```python
document.tables[5].columns[0].cells[0].text
```

```python
#All tables:
usecase = []
```

```python
# r.cells[0].text
```

```python
t = 0
for t in document.tables[:20]:
    if t.cell(0,0).text == "Use case id":
        print(f"Use Case: {t.cell(0,1).text}")
        use_case = {}
        for r in t.rows:
            use_case[r.cells[0].text] = r.cells[1].text
        usecase.append(use_case)
        print("Table is Parsed well")
```

```python
import pandas as pd
usecase_df = pd.DataFrame()
```

```python
for u in usecase:
    new_u = {}
    for key in u:
        new_u[key] = [u[key]]
    new_u = pd.DataFrame(new_u)
    usecase_df = pd.concat([usecase_df, new_u], ignore_index = True)
```

```python
usecase_df
```

```python
usecase_df.loc[usecase_df["Use case description"].isna(), "Use case description"] = usecase_df.loc[usecase_df["Use case description"].isna(), "Description"]
```

```python
usecase_df = usecase_df.drop(columns = ["Description"])
```

```python
usecase_df.loc[usecase_df["Open issues"].isna(), "Open issues"] = usecase_df.loc[usecase_df["Open issues"].isna(), "Open issues (as seen by the proponent)"]
```

```python
usecase_df = usecase_df.drop(columns = ["Open issues (as seen by the proponent)"])
```

```python
usecase_df.loc[usecase_df["Category"].isna(), "Category"] = usecase_df.loc[usecase_df["Category"].isna(), "Notes on use case category"]
```

```python
usecase_df = usecase_df.drop(columns = ["Notes on use case category"])
```

```python
usecase_df.loc[usecase_df["Use case category"].isna(), "Use case category"] = usecase_df.loc[usecase_df["Use case category"].isna(), "Category"]
```

```python
usecase_df = usecase_df.drop(columns = ["Category"])
```

```python
usecase_df.to_csv("usecases.csv", index = False)
```
